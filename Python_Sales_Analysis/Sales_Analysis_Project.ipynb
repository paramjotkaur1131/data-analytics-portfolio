```python
# Import essential libraries
import pandas as pd   #For handling data
import numpy as np    #For numerical calculations and working with arrays
import matplotlib.pyplot as plt   #For making charts
#Load your Superstore CSV file
data = pd.read_csv('/content/Superstore Sales Dataset.csv')
data.head()
```

```python
data.info()
```

```python
# Check if any values are missing
print("Missing values in each column:")
print(data.isnull().sum())

# Remove duplicate rows (if any)
data.drop_duplicates(inplace=True)
```

```python
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()

print("Total Sales:", round(total_sales, 2))
print("Total Profit:", round(total_profit, 2))
```

```python
region_sales = data.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
category_sales = data.groupby('Category')['Sales'].sum()
print(category_sales)
```

```python
# Sales by Region
region_sales.plot(kind='bar', title='Sales by Region', color='skyblue')
plt.ylabel('Total Sales')
plt.show()

# Sales by Category
category_sales.plot(kind='bar', title='Sales by Category', color='lightgreen')
plt.ylabel('Total Sales')
plt.show()
```
